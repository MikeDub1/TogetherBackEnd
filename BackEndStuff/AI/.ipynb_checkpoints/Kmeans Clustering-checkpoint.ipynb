```python
#import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import pandas as pd
import plotly.express as px
df = pd.read_csv (r'new_dataframe.csv')
```

```python
clustering_kmeans = KMeans(n_clusters=5, precompute_distances="auto", n_jobs=-1)
df['clusters'] = clustering_kmeans.fit_predict(df)
```

```python
df.clusters
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df.loc[df['clusters'] == 0]
```

```python
df.describe().transpose()
```

```python
fig = px.scatter_matrix(df.drop("clusters",axis=1),
width=1200, height=1600)
fig.show()
```

```python
fig2 = px.scatter_3d(df, x="books", y="Games",z="travel",
                     color="Cooking",size="GYM")
fig2.update_layout(title="5 Features Representation")
fig2.show()
```

```python

from sklearn.preprocessing import MinMaxScaler
import plotly.graph_objects as go
X=df.drop("clusters",axis=1)
scaler = MinMaxScaler()
scaler.fit(X)
X=scaler.transform(X)
inertia = []
for i in range(1,11):
    kmeans = KMeans(
        n_clusters=i, init="k-means++",
        n_init=10,
        tol=1e-04, random_state=42
    )
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
fig = go.Figure(data=go.Scatter(x=np.arange(1,11),y=inertia))
fig.update_layout(title="Inertia vs Cluster Number",xaxis=dict(range=[0,11],title="Cluster Number"),
                  yaxis={'title':'Inertia'},
                 annotations=[
        dict(
            x=3,
            y=inertia[2],
            xref="x",
            yref="y",
            text="Elbow!",
            showarrow=True,
            arrowhead=7,
            ax=20,
            ay=-40
        )
    ])
```
